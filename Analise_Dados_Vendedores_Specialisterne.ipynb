Criando a Função para o Tratamento dos Dados

```python
def tratar_dataframe(df):
    # Definir o nome correto das colunas
    colunas = [
        'Unnamed_0', 'Data', 'Unnamed_2', 'Venda_Real', 'Venda_Projetada',
        'Diferença', 'Unnamed_6', 'Acumulado_Real', 'Acumulado_Projetado',
        'Unnamed_9', 'Unnamed_10', 'Tendencia','Unnamed_12','Clientes_Abordados','Vendas_Fechadas','Taxa_Conversao'
           ]

    # Aplicar as colunas ao dataframe
    df.columns = colunas

    # Remover colunas desnecessárias
    df = df.drop(columns=['Unnamed_0', 'Unnamed_2', 'Unnamed_6', 'Unnamed_9', 'Unnamed_10','Unnamed_12'])

    # Remover as primeiras linhas irrelevantes (ajustado para remover apenas o que for necessário)
    df = df[5:].reset_index(drop=True)

    # Converter a coluna Data para o formato datetime
    df['Data'] = pd.to_datetime(df['Data'], format='%Y-%m-%d', errors='coerce')

    # Verificar e limpar colunas numéricas (remover caracteres indesejados)
    colunas_numericas = ['Venda_Real', 'Venda_Projetada', 'Diferença', 'Acumulado_Real',
                         'Acumulado_Projetado', 'Clientes_Abordados', 'Vendas_Fechadas',
                         'Taxa_Conversao']

    df[colunas_numericas] = df[colunas_numericas].apply(pd.to_numeric, errors='coerce')

    # Remover a primeira e a última linha
    df = df.iloc[1:-1]

    return df

# Aplicar a função aos dataframes df_ana, df_joao e df_jose
df_ana = tratar_dataframe(df_ana)
df_joao = tratar_dataframe(df_joao)
df_jose = tratar_dataframe(df_jose)

# Exibir os primeiros dados limpos para verificar o resultado
print(df_ana.head())
print(df_joao.head())
print(df_jose.head())
```

## **Função Análise de Vendedores**

**Métricas de Desempenho:**
1. Total de Vendas (Soma de todas as vendas do mês)
2. Taxa de Conversão (Taxa média da divisão entre Quantidade de Vendas Fechadas / Clientes Abordados)
3. Diferença Total de Vendas Projetadas - Real (Taxa média. Se o valor for positivo significa dizer que foi vendido mais que o esperado)
4. Meta Mensal (Se o vendedor atingiu a meta de R$ 2000,00)


```python
import pandas as pd
import matplotlib.pyplot as plt

def analisar_vendedor(df, nome_vendedor, meta_mensal=2000):
    """
    Analisa o desempenho de um vendedor.

    Args:
        df: DataFrame com os dados de vendas do vendedor.
        nome_vendedor: Nome do vendedor.
        meta_mensal: Meta de vendas mensal.

    """

    # Cálculo das métricas
    total_vendas = df['Venda_Real'].sum()
    taxa_conversao = (df['Vendas_Fechadas'] / df['Clientes_Abordados']).mean() * 100
    diferenca_media = df['Diferença'].mean()

    # Criar o gráfico
    sns.set_style("whitegrid")  # Define o estilo do gráfico
    plt.figure(figsize=(12, 6))
    ax = sns.lineplot(data=df, x='Data', y='Acumulado_Real', label='Acumulado Real', marker='o')
    sns.lineplot(data=df, x='Data', y='Acumulado_Projetado', label='Acumulado Projetado', linestyle='--')

    # Formatação do gráfico
    ax.set(xlabel='Data', ylabel='Valor Acumulado')
    ax.set_title(f'Progresso Acumulado das Vendas - {nome_vendedor}', size=20, loc='left', weight='bold')
    ax.tick_params(axis='x', rotation=30)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: f'R$ {int(x):,}'.replace(',', '.')))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%y'))


    plt.show()

    # Impressão dos resultados
    print(f"Relatório para {nome_vendedor}:")
    print(f"Total de Vendas: R$ {total_vendas:.2f}")
    print(f"Taxa de Conversão: {taxa_conversao:.2f}%")
    print(f"Diferença Média entre Venda Real e Projetada: R$ {diferenca_media:.2f}")
    print(f"Atingiu a meta de R$ {meta_mensal}? {'Sim' if total_vendas >= meta_mensal else 'Não'}")

# Aplicar a função aos dataframes
if __name__ == "__main__":
    # Certifique-se que df_ana, df_joao e df_jose estão definidos corretamente
    # ... (seu código para carregar os dados)

    analisar_vendedor(df_ana, 'Ana')
    analisar_vendedor(df_joao, 'João')
    analisar_vendedor(df_jose, 'José')
```

```python
# Criar um gráfico de linhas para todos os vendedores
def plotar_comparacao_progressao(df_ana, df_joao, df_jose):
    plt.figure(figsize=(10, 6))
    for vendedor, df in zip(['Ana', 'João', 'José'], [df_ana, df_joao, df_jose]):
        plt.plot(df['Data'], df['Acumulado_Real'], label=vendedor)

    plt.xlabel(' ')
    plt.ylabel('Valor Acumulado (R$)')
    plt.title('Comparação da Progressão das Vendas', size=20)
    plt.legend()
    plt.grid(True)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%y'))
    plt.show()

# Chamar a função
plotar_comparacao_progressao(df_ana, df_joao, df_jose)
```

Gráfico relacionado ao valor real de venda acumulado com o período selecionado.

```python
import matplotlib.pyplot as plt

def plotar_comparativo_vendas(df_ana, df_joao, df_jose):
    """
    Cria um gráfico de barras comparando o total de vendas de cada vendedor.
    """

    total_vendas = [df_ana['Venda_Real'].sum(), df_joao['Venda_Real'].sum(), df_jose['Venda_Real'].sum()]
    vendedores = ['Ana', 'João', 'José']

    plt.figure(figsize=(8, 5))
    colors = ['blue', 'orange', 'green']  # Cores para o gráfico
    bars = plt.bar(vendedores, total_vendas, color=colors)

    plt.xlabel(' ')
    plt.ylabel('Total de Vendas (R$)')
    plt.title('Comparação do Total de Vendas', size=20)

    # Adicionando rótulos nas barras
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, f'R$ {yval:.2f}', ha='center', va='bottom')

    plt.legend(['Ana', 'João', 'José'])
    plt.show()

# Chamar a função para plotar o gráfico
plotar_comparativo_vendas(df_ana, df_joao, df_jose)
```

Comparação do valor total de vendas de cada vendedor no período estimado.

```python
import matplotlib.pyplot as plt

def plotar_histograma_vendas(df_ana, df_joao, df_jose):
    """
    Cria um histograma comparando a distribuição das vendas de cada vendedor.
    """

    plt.figure(figsize=(10, 6))
    plt.hist([df_ana['Venda_Real'], df_joao['Venda_Real'], df_jose['Venda_Real']],
             bins=10,
             label=['Ana', 'João', 'José'],
             color=['blue', 'orange', 'green'],
             alpha=0.7)

    plt.xlabel('Valor das Vendas (R$)')
    plt.ylabel('Frequência')
    plt.title('Distribuição das Vendas por Vendedor', size=20)
    plt.legend()

    # Remover a grade
    plt.grid(False)  # Desabilitar a grade

    plt.show()

# Chamar a função para plotar o histograma
plotar_histograma_vendas(df_ana, df_joao, df_jose)
```
